```python
def rock_paper_game():
    while(True):
        print("You are Enter To ROCK PAPER SCISSOR GAME")
        import random as rd
        user=input("Select any among using there corresponding numbers \nROCK\nSCISSOR \nPAPER").lower()
        while user not in ["rock","paper","scissor"]:
            print("invalid choice \n choose again")
            user = input("Select any among the following: ROCK, PAPER, SCISSOR\n").lower()
        all_moves=["rock","paper","scissor"]
        computer=rd.choice(all_moves)
        print("computer select ",computer)
        if user == computer:
            print("TIE")
        elif  (user == 'rock' and computer == 'scissors') or \
                 (user == 'scissors' and computer == 'paper') or \
                 (user == 'paper' and computer == 'rock'):
            print("user win")
        else:
            print("computer win")
        replay=input("Do YOU want to start again the game (y/n)").lower()
        if replay=="n":
            break;
        else:
            print("Entr to game again congs")
rock_paper_game()
```
